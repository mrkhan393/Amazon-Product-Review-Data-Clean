# Normalize text fields

## Check text to be normalized

```python
def check_whitespace(df, column):
    leading_whitespace = df[column].str.startswith(' ')
    trailing_whitespace = df[column].str.endswith(' ')
    return leading_whitespace.any() or trailing_whitespace.any()
```

```python
columns_to_check = ['user_name', 'product_name']

for column in columns_to_check:
    if check_whitespace(df, column):
        print(f"column '{column}' contains leading or trailing whitespace.")
    else:
        print(f"column '{column}' does not contain leading or trailing whitespace.")
```

# Split user id, user name, review id, review title

```python
# Columns to be split
split_columns = ['user_id', 'user_name', 'review_id', 'review_title']

# Splitting the specified columns by commas and replacing None with empty lists
split_data = {col: df[col].str.split(',').apply(lambda x: x if x is not None else []) for col in split_columns}

# Finding the maximum length after splitting to pad other lists
max_length = max([max(split_data[col].apply(len)) for col in split_columns])
```

```python
# Padding lists with None where necessary to ensure equal lengths
for col in split_columns:
    split_data[col] = split_data[col].apply(lambda x: x + [None] * (max_length - len(x)))
```

```python
# Creating a new DataFrame from the split data
expanded_df = pd.DataFrame({col: [item for sublist in split_data[col] for item in sublist] for col in split_columns})

# Dropping the split columns from the original DataFrame
other_columns = df.drop(columns=split_columns)

# Expanding other columns to match the length of the expanded split columns
other_columns = other_columns.loc[other_columns.index.repeat(max_length)].reset_index(drop=True)
```

```python
# Combining the expanded columns with the other columns
df = pd.concat([other_columns, expanded_df], axis=1)

# Dropping rows where all the split columns have None values
df = df.dropna(subset=split_columns, how='all').reset_index(drop=True)

# Show the resulting DataFrame
df.tail(50)
```
